```python
#These columns are being dropped because had over 140,000 missing data values.
dropped_columns = ['benchPoints', 'timeoutsRemaining', 'leadChanges', 'biggestScoringRun', 'biggestLead',
                  'pointsFastBreak', 'pointsSecondChance', 'pointsInThePaint', 'pointsFromTurnovers', 'seasonLosses',
                  'seasonWins', 'timesTied', 'q1Points', 'q2Points', 'q4Points', 'q3Points', 'coachId']
nba = nba.drop(columns=dropped_columns)
```
